```python
import os
import numpy as np
from sklearn.mixture import GaussianMixture
from scipy.ndimage import gaussian_filter, binary_fill_holes
import surfa as sf
import param_3d  # your module
from utils import shift_non_zero_elements  # your utility

def build_gmm_label_map_combined(k1=5, k2=6):  # k1=5 classes for brain, label 6 is fallback
    folders_path = [
        "/gpfs/fs001/cbica/home/dadashkj/upenn_pigAnatomical/81-3day",
        "/gpfs/fs001/cbica/home/dadashkj/upenn_pigAnatomical/81-3day-T2"
    ]

    def load_volume(folder, name):
        for ext in ['.nii.gz', '.nii']:
            path = os.path.join(folder, name + ext)
            if os.path.exists(path):
                return sf.load_volume(path).reshape((param_3d.img_size_256,) * 3).data
        raise FileNotFoundError(f"Missing {name} in {folder}")

    predicted_anat_labels = []
    sigma = 0.8
    i = 0

    while i < len(folders_path):
        t1_folder = folders_path[i]
        t2_folder = folders_path[i + 1] if i + 1 < len(folders_path) else None
        has_t2 = t2_folder and "T2" in t2_folder and t1_folder.replace("-T2", "") in t2_folder

        anat_t1 = load_volume(t1_folder, "anat")
        mask = load_volume(t1_folder, "anat_brain_olfactory_mask") > 0
        mask = binary_fill_holes(mask).astype(np.uint8)

        anat_t1 = gaussian_filter(anat_t1, sigma)

        if has_t2:
            anat_t2 = load_volume(t2_folder, "anat")
            anat_t2 = gaussian_filter(anat_t2, sigma)
            anat_combined = np.stack([anat_t1, anat_t2], axis=-1)
            brain_data = anat_combined[mask == 1].reshape(-1, 2)
            non_brain_data = anat_combined[mask == 0].reshape(-1, 2)
        else:
            brain_data = anat_t1[mask == 1].reshape(-1, 1)
            non_brain_data = anat_t1[mask == 0].reshape(-1, 1)

        # GMM Fitting
        gmm_brain = GaussianMixture(n_components=k1, random_state=42).fit(brain_data)
        gmm_non_brain = GaussianMixture(n_components=k2, random_state=42).fit(non_brain_data)

        # Initialize full_seg with zeros (background)
        full_seg = np.zeros(mask.size, dtype=int)

        # Predict brain voxels (1â€“k1), assign 6 to failed ones
        brain_idx = mask.flatten() == 1
        brain_pred = gmm_brain.predict(brain_data)

        full_seg[brain_idx] = 6  # fallback class
        if len(brain_pred) > 0:
            full_seg[brain_idx] = np.where(brain_pred < k1, brain_pred + 1, 6)

        # Predict background (k1+1 to k1+k2)
        non_brain_idx = ~brain_idx
        non_brain_pred = gmm_non_brain.predict(non_brain_data)
        full_seg[non_brain_idx] = non_brain_pred + k1 + 1

        # Finalize
        full_seg = full_seg.reshape((param_3d.img_size_256,) * 3)
        full_seg = shift_non_zero_elements(full_seg, 1)

        predicted_anat_labels.append(sf.Volume(full_seg).reshape((param_3d.img_size_256,) * 3))

        i += 2 if has_t2 else 1

    return predicted_anat_labels

# Run
a = build_gmm_label_map_combined()
```

```python
import os
import numpy as np
from sklearn.mixture import GaussianMixture
from scipy.ndimage import gaussian_filter, binary_fill_holes
import surfa as sf
import param_3d  # your module
from utils import shift_non_zero_elements  # your utility

def build_gmm_label_map_combined(k1=5, k2=6):
    t1_folder = "/gpfs/fs001/cbica/home/dadashkj/upenn_pigAnatomical/81-3day"
    t2_folder = "/gpfs/fs001/cbica/home/dadashkj/upenn_pigAnatomical/81-3day-T2"

    def load_volume(folder, name):
        for ext in ['.nii.gz', '.nii']:
            path = os.path.join(folder, name + ext)
            if os.path.exists(path):
                return sf.load_volume(path).reshape((param_3d.img_size_256,) * 3).data
        raise FileNotFoundError(f"Missing {name} in {folder}")

    def segment_gmm(image_data, mask, k1, k2):
        brain_data = image_data[mask == 1].reshape(-1, image_data.shape[-1])
        non_brain_data = image_data[mask == 0].reshape(-1, image_data.shape[-1])

        gmm_brain = GaussianMixture(n_components=k1, random_state=42).fit(brain_data)
        gmm_non_brain = GaussianMixture(n_components=k2, random_state=42).fit(non_brain_data)

        full_seg = np.zeros(mask.size, dtype=int)

        brain_idx = mask.flatten() == 1
        brain_pred = gmm_brain.predict(brain_data)
        full_seg[brain_idx] = np.where(brain_pred < k1, brain_pred + 1, 6)

        non_brain_idx = ~brain_idx
        non_brain_pred = gmm_non_brain.predict(non_brain_data)
        full_seg[non_brain_idx] = non_brain_pred + k1 + 1

        full_seg = full_seg.reshape((param_3d.img_size_256,) * 3)
        full_seg = shift_non_zero_elements(full_seg, 1)
        return sf.Volume(full_seg).reshape((param_3d.img_size_256,) * 3)

    # Load T1, T2, and mask
    anat_t1 = gaussian_filter(load_volume(t1_folder, "anat"), 0.8)
    anat_t2 = gaussian_filter(load_volume(t2_folder, "anat"), 0.8)
    mask = load_volume(t1_folder, "anat_brain_olfactory_mask") > 0
    mask = binary_fill_holes(mask).astype(np.uint8)

    # Segmentations
    seg_t1 = segment_gmm(anat_t1[..., np.newaxis], mask, k1, k2)
    seg_t2 = segment_gmm(anat_t2[..., np.newaxis], mask, k1, k2)
    seg_combined = segment_gmm(np.stack([anat_t1, anat_t2], axis=-1), mask, k1, k2)

    return seg_t1, seg_t2, seg_combined

# Run
seg_t1, seg_t2, seg_combined = build_gmm_label_map_combined()
```
